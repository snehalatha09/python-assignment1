#maximum of two numbers

```
def max(a,b,c):
  if (a>b) and (a>c):
    print("a is max")
  elif (b>a) and (b>c):
    print(" b is max")
  else:
    print("c is max",c)
  return max 
print(max(1,2,3))
```

#factorial of a number

```
def fact(num):
  if num==1:
    return num
  else:
    return num*fact(num-1)
print((fact(7)))
```
